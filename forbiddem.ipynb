```python
#importing all the required libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.stem import WordNetLemmatizer
from nltk import tokenize
```

```python
#converting the reviews into lower case
df.review=df.review.apply(lambda x: x.lower())
```

```python
# Expanding contractions
import contractions
def con(text):
    expand=contractions.fix(text)
    return expand

df.review=df.review.apply(con)
df['review'][0]
```

```python
import re

def remove_sp(text):
    pattern=r'[^A-Za-z0-9\s]'
    text=re.sub(pattern,'',text)
    return text

df.review=df.review.apply(remove_sp)
df
```

```python
import string
punctuations=list(string.punctuation)
```

```python
df.review=df.review.apply(lambda x : " ".join(x for x in x.split() if x not in punctuations))
```

```python
# Removing stopwords
nltk.download('stopwords')
stopword_list=stopwords.words('english')
```

```python
stopword_list.remove('no')
stopword_list.remove('not')
```

```python
df.review=df.review.apply(lambda x : " ".join(x for x in x.split() if x not in stopword_list))
df['review'][5]
```

```python
nltk.download('punkt')
```

```python
df['review']=df.review.apply(word_tokenize)
df['review'][0]
```

```python
nltk.download('wordnet')
```

```python
lemmatizer=WordNetLemmatizer()
```

```python

df['review']=df.review.apply(lambda x:[lemmatizer.lemmatize(word) for word in x])
```

```python
df.review= df.review.astype(str)
```

```python
import numpy as np
review_list=df['review'].array
```

```python
x = df.iloc[:,1].values
y = df.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
```

```python
final = Pipeline([('Vect',TfidfVectorizer()),
                  ('model',SVC(probability=True))])
```

```python
final.fit(x_train,y_train)
```

```python
def Sentiments(Review):
  x.apply(lambda x: x.lower())
  x.apply(con)
  x.apply(remove_sp)
  x.apply(lambda x : " ".join(x for x in x.split() if x not in punctuations))
  x.apply(lambda x : " ".join(x for x in x.split() if x not in stopword_list))
  x.apply(word_tokenize)
  x.apply(lambda x:[lemmatizer.lemmatize(word) for word in x])
  x.astype(str)
  prediction = final.predict(x)
  return prediction
  #return x

```
