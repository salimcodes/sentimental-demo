```python
#importing all the required libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.stem import WordNetLemmatizer
from nltk import tokenize
```

```python
#converting the reviews into lower case
df.review=df.review.apply(lambda x: x.lower())
```

```python
# Expanding contractions
import contractions
def con(text):
    expand=contractions.fix(text)
    return expand

df.review=df.review.apply(con)
df['review'][0]
```

```python
import re

def remove_sp(text):
    pattern=r'[^A-Za-z0-9\s]'
    text=re.sub(pattern,'',text)
    return text

df.review=df.review.apply(remove_sp)
df
```

```python
import string
punctuations=list(string.punctuation)
```

```python
df.review=df.review.apply(lambda x : " ".join(x for x in x.split() if x not in punctuations))
```

```python
# Removing stopwords
nltk.download('stopwords')
stopword_list=stopwords.words('english')
```

```python
stopword_list.remove('no')
stopword_list.remove('not')
```

```python
df.review=df.review.apply(lambda x : " ".join(x for x in x.split() if x not in stopword_list))
df['review'][5]
```

```python
nltk.download('punkt')
```

```python
df['review']=df.review.apply(word_tokenize)
df['review'][0]
```

```python
nltk.download('wordnet')
```

```python
lemmatizer=WordNetLemmatizer()
```

```python

df['review']=df.review.apply(lambda x:[lemmatizer.lemmatize(word) for word in x])
```

```python
import gradio as gr

# Assuming the previous code is already present

# Initialize the pipeline (from the previous code)
final = Pipeline([('Vect', TfidfVectorizer()), ('model', SVC(probability=True))])
final.fit(x_train, y_train)

# Define the prediction function using the trained pipeline
def predict_sentiment(review):
    result = final.predict([review])[0]
    return result

# Create a Gradio interface
iface = gr.Interface(
    fn=predict_sentiment,
    inputs=gr.inputs.Textbox(lines=5, placeholder="Enter your review here..."),
    outputs=gr.outputs.Label(),
    title="Sentiment Analysis App"
)

# Launch the Gradio interface
iface.launch()
```
